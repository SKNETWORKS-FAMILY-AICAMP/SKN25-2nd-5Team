```python
drop_columns = [
    '직원ID',
    '직원번호',
    '직원수',
    '18세이상여부',
    '표준근무시간'
]


df = df.drop(columns=drop_columns)
```

```python
df = df.drop(columns=['시급', '급여구간', '월별지급액'])
df = df.drop(columns='일급')
df = df.drop(columns='연령대')
```

```python
job_mapping = {
    "Healthcare Representative": "전문직",
    "Human Resources": "인사직",
    "Laboratory Technician": "연구/기술직",
    "Research Scientist": "연구/기술직",
    "Research Director": "연구/관리직",
    "Manufacturing Director": "생산/관리직",
    "Manager": "관리직",
    "Sales Executive": "영업직",
    "Sales Representative": "영업직"
}

df["직무분류"] = df["직무"].map(job_mapping)

df = df.drop(columns=["직무"])

df["직무분류"].value_counts()
```

```python
df.groupby("퇴사여부")["전공분야"].value_counts(normalize=True)
```

```python
department_mapping = {
    "Sales": "영업부",
    "Research & Development": "연구개발부",
    "Human Resources": "인사부"
}

df["부서"] = df["부서"].map(department_mapping)

# 확인
df["부서"].value_counts()
```
